```python
# Import necessary libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
# Load the Wine dataset
wine = datasets.load_wine()
```

```python
# Split the data into training and testing sets
x = wine.data
y = wine.target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
# Support Vector Machine (SVM)
svm_model = SVC()
svm_model.fit(x_train, y_train)
svm_pred = svm_model.predict(x_test)
svm_accuracy = metrics.accuracy_score(y_test, svm_pred)
```

```python
# Decision Tree
dt_model = DecisionTreeClassifier()
dt_model.fit(x_train, y_train)
dt_pred = dt_model.predict(x_test)
dt_accuracy = metrics.accuracy_score(y_test, dt_pred)
```

```python
# Naive Bayes
nb_model = GaussianNB()
nb_model.fit(x_train, y_train)
nb_pred = nb_model.predict(x_test)
nb_accuracy = metrics.accuracy_score(y_test, nb_pred)
```

```python
# Display accuracies
print("SVM Accuracy:",svm_accuracy)
print("Decision Tree Accuracy:",dt_accuracy)
print("Naive Bayes Accuracy:",nb_accuracy)
```

```python
# Plotting
classifiers = ['SVM', 'Decision Tree', 'Naive Bayes']
accuracies = [svm_accuracy, dt_accuracy, nb_accuracy]

plt.bar(classifiers, accuracies, color=['Blue', 'Red', 'Pink'])
plt.xlabel('Classifiers')
plt.ylabel('Accuracy')
plt.title('Classifier Comparison on Wine Dataset')
plt.show()
```
